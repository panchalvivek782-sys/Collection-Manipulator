```python
students={}
print("Welcome to the Student Data Organizer!")

while True:
    print("\nSelect an option:")
    print("1. Add Student")
    print("2. Display All Students")
    print("3. Update Student Information")
    print("4. Delete Student")
    print("5. Display Subjects Offered")
    print("6. Exit")

    choice=int(input("Enter your choice: "))
    if choice==1:
        print("\nEnter Student details:")
        student_id=input("Enter Student ID: ").strip()

        if student_id in students:
            print("Error: A student with this ID already exists.")
            continue

        name=input("Name: ")
        age=int(input("Age: "))
        grade=input("Grade: ")
        dob=input("Date of Birth (YYYY-MM-DD): ")
        subjects_input=input("Subjects (comma-separated): ")

        subjects=[s for s in subjects_input.split(",") if s]

        students[student_id]={"name":name,"age":age,"grade":grade,"dob":dob,"subjects":subjects}
    elif choice==2:
        print("\n--- Display All Students ---")
        if not students:
            print("No student records found.")
        else:
            for sid, info in students.items():
                subjects_str=", ".join(info["subjects"])
                print(f"Student ID: {sid} | Name: {info["name"]} | Age: {info["age"]} | DOB:{info["dob"]} |"
                      f"Grade: {info["grade"]} | Subjects: {subjects_str}")
    elif choice==3:
        print("\n--- Update Student Information ---")
        student_id=input("Enter Student ID to update: ")
        if student_id in students:
            print("Leave blank to keep the current value.")
            name=input(f"New Name ({students[student_id]["name"]}): ")
            if name: students[student_id]["name"]=name

            age=input(f"New Age ({students[student_id]["age"]}): ")
            if age: students[student_id]["age"]=age

            grade=input(f"New Grade ({students[student_id]["grade"]}): ")
            if grade: students[student_id]["grade"]=grade

            dob=input(f"New DOB ({students[student_id]["dob"]}): ")
            if dob: students[student_id]["dob"]=dob

            current_subjects=", ".join(students[student_id]["subjects"])
            subjects_input=input(f"New Subjects ({current_subjects}): ")
            if subjects_input:
                students[student_id]["subjects"]=[s for s in subjects_input.split(",") if s]

            print("Students information updated successfully!")
        else:
            print("Student ID not found.")

    elif choice==4:
        print("\n--- Delete Student ---")
        student_id=input("Enter Student ID to delete: ")
        if student_id in students:
            del students[student_id]
            print("Student record deleted successfully!")
        else:
            print("Student ID not found.")

    elif choice==5:
        print("\n--- Display Subjects Offered ---")
        unique_subjects=[]
        for info in students.values():
            for subject in info["subjects"]:
                if subject not in unique_subjects:
                    unique_subjects.append(subject)

        if unique_subjects:
            sorted_subjects=sorted(unique_subjects)
            print("Subjects currently offered:", ", ".join(sorted_subjects))
        else:
            print("No subjects are currently registered by any student.")

    elif choice==6:
        print("Exiting Student Data Organizer. Goodbye!")
        break

    else:
        print("Invalid choice! Please enter a number between 1 and 6.")
```
